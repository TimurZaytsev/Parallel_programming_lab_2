```python
time_p = [8.64393, 4.17985, 2.99085, 2.19956, 1.66312]
time_l = [0.772841, 0.33579, 0.217967, 0.163434, 0.134197]
delta = [0.001, 0.002, 0.003, 0.004, 0.005]
```

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

plt.plot(delta, time_p, label = "Параллельное")
plt.plot(delta, time_l, label = "Линейное")
plt.xlabel("delta T")
plt.ylabel("Время выполнения")
plt.legend()
plt.show()
```

```python
time = [time_l[i] / time_p[i] for i in range(5)]
plt.plot(delta, time, label = "Ускорение")
plt.xlabel("delta T")
plt.ylabel("Время выполнения")
plt.legend()
plt.show()
```
